```sql
/*
(10p) Soru 1 : Her bir siparişin kaç parça(item) içerdiğini aşağıdaki şekilde listeleyen kodu yazınız.
             - sipariş ID ve adet bilgisini( as countOfItems) içermelidir
             - sırasıyla adet bilgisi ve sipariş ID 'ye göre büyükten küçüğe doğru olmalıdır
*/  
use BikeStores
select sum(item_id) as countOfItems, order_id  
from order_items
group by order_id
order by order_id desc

/* 
use ile veri tabanı bağlantısı kuruldu.
Sum ile parçaların toplam ne kadar gerektiği hesaplandı. order_id ile id'ye göre ne kadar gerektiği yazdırıldı.
from ile bu yazdırmalar order_items tablosundan çekildi.
group by ile order_id'ye göre gruplama yaparak order by ile büyükten küçüğe göre sıraladım.
*/
```

```sql
/*
(10p) Soru 2: En çok satılan(order_items) ilt 5 ürünü(product) aşağıdaki şekilde listeleyen kodu yazınız.
            - ürün ID ve adet sayısını(as countOfOrder) içermelidir.
            - adet sayısına göre büyükten küçüğe doğru sıralanmalıdır
            - uygun JOIN yöntem-leri kullanılmalıdır !!
*/
USE BikeStores
SELECT TOP 5 products.product_id, count (order_items.quantity) as countOfOrder
from products
inner join order_items on products.product_id=order_items.product_id 
group by products.product_id
order by count(order_items.quantity) desc

/* 
    use kullanılarak veritabanı bağlantısı kurulur.
    En çok satılan ilk 5 ürün dediği için top 5 yazıldı. en çok kısmı ise birazdan girilmiş olacak. Count ile miktar belirtildi.
    from ile ürünlerden çekildi.
    join on bağlantısı ile order_items ile products arasında bağlantı kuruldu. Böylece order_items.quantity ürün id'ye göre çekilmiş olundu.
    En sonda group by eklenerekgruplama yapılarak order by ile de büyükten küçüğe kısmı yapıldı. top 5 dediği için de ilk 5 ürün çekilmiş olundu.
*/
```

```sql
/*
(10p) Soru 3 : Toplam indirim(discount) miktarı 2'den fazla olan ürünleri(product) aşağıdaki şekilde listeleyen kodu yazınız.
             - ürün ID ve toplam indirim bilgisini( as sumOfDiscount) içermelidir
             - ürün ID'ye göre küçükten büyüğe doğru olmalıdır
             - uygun JOIN yöntem-leri kullanılmalıdır !!
*/
use BikeStores
select products.product_id, sum(order_items.discount) as sumOfDiscount
from products
join order_items on products.product_id = order_items.product_id
group by products.product_id
having sum(order_items.discount) > 2
order by products.product_id asc

/* 
    use ile veritabanı bağlantısı kuruldu
    select ile id'ye göre indirim miktarı 2'den fazla olan ürünler çekildi.
    bunu da from ile products'dan yaptım.
    ardından join on ile order_items ile products arasında bağı kurdum.
    group by ile gruplama yaparak having sum ile 2'den fazla olanların şartının sorgusunu yaptım.
    order by ile küçükten büyüğe işlemini yaptım.
*/
```

```sql
/*
(15p) Soru 5: Her bir mağaza için yapılan toplam indirimi(discount) aşağıdaki şekilde listeleyen kodu yazınız.
             - mağaza ID ve toplam indirim bilgisini( as sumOfDiscount) içermelidir
             - uygun JOIN yöntem-leri kullanılmalıdır !!
*/
USE BikeStores
select stores.store_id, sum(order_items.discount) as sumOfDiscount
from stores
join orders on orders.store_id = stores.store_id
join order_items on order_items.order_id = orders.order_id
group by stores.store_id



/* 
use ile veritabanı bağlantısı yapıldı.
select ile mağazaya göre toplam indirim sayıları yazıldı. bunun için de 
from ile stores dan çekildi veriler ilk önce
join on ile stores ile orders arasında ki bağ yazıldı.
ardından join on ile order ile order_items arasında bağ yapıldı. Böylece siparişler ile  mağzalar ve siparişler ile itemslar arasında bağ kurulmuş oldu
Böylece çözülmüş oldu soru. Şimdi group by ile store_id' tekrarını önledikten sonra hazır.
*/
```

```sql
/*
(20p) Soru 6: Her bir mağaza ve sipariş(order) için toplam parça(item) sayısını aşağıdaki şekilde listeleyen kodu yazınız.
             - mağaza ID, sipariş ID ve toplam ürün sayısını( as countOfItems) içermelidir
             - sırasıyla mağaza ID, sipariş ID ve toplam ürün sayısına göre küçükten büyüğe doğru sıralanmalıdır
             - uygun JOIN yöntem-leri kullanılmalıdır !!
*/
USE BikeStores 
select stores.store_id, orders.order_id, sum(order_items.quantity) as countOfItems
from stores
join orders on orders.store_id = stores.store_id
join order_items on order_items.order_id = orders.order_id
group by stores.store_id, orders.order_id
order by stores.store_id asc, orders.order_id asc, count(order_items.quantity)asc

/*      
    use ile veri bağlantısı kuruldu.
    select ile mağaza id, sipariş id ve toplam ürün sayısı basıldı.
    bunları yapabilmem için ilk önce from ile mağazalardan verileri çektim ardından join on ile orders ile stores ve order_items ile orders arasında ki bağı kurdum.
    Böylece sırasıyla mağaza id, sipariş id ve toplam ürün sayısını çekebilirim. Ancak sırasıyla çekebilmem için de küçükten büyüğe order by'da o şekilde yazmam gerekiyordu.
    asc ile de küçükten büyüğe doğru sıraladım. group by ile veri tekrarının önüne geçtim. Gruplamayı da bu şekilde yaparak tamamladım.
*/
```

```sql
/*
(20p) Soru 7: Toplam stok sayısı(quantity) 25'ten fazla olan ürünleri(product) aşağıdaki şekilde listeleyen kodu yazınız.
            - ürün ID,ürün adı, marka ID, model yılı ve stok sayısı(as sumOfQuantity)
            - ürün adına göre küçükten büyüğe doğru sıralı olmalısır
            - uygun JOIN yöntem-leri ve SELECT yapıları kullanılmalıdır !!
*/
use BikeStores
select products.product_id, products.product_name, products.brand_id, products.model_year, sum(stocks.quantity) as sumOfQuantity
from products
join stocks on products.product_id = stocks.product_id
group by products.product_id, products.product_name, products.brand_id, products.model_year
having sum(stocks.quantity) > 25
order by products.product_name asc

/* 
Toplam stok sayısı having ile koşul komut satırı yazıldı.
from ile ürünlerden veri çekilerek sonuçlar yazdırıldı.
sırasıyla select ile ürün id, ürün adı, marka id, model yılı ve stok sayısı yazdırılmış olundu. bunlar aynı sıra ile
group by ile de yazdırılarak gruplama yapılmasını sağladım.
ürün adına göre küçükten büyüpe dorğu yazılması asc ile order by kullanımında olmaktadır.
join on kullanımı ile de stocks ve products arasında bağ kurulmuş oldu.

    
*/
```

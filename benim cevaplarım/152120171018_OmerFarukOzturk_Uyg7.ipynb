```sql
/*
(10p) Soru 1 : Adı "b", "s" veya "p" ile başlamayan ürünleri(products) listeleyiniz.
               * Bütün sütünlar listelenebilir.
*/

use BikeStores
select * from products
where product_name not like 'b%' and product_name not like 's%' and product_name not like 'p%'  


/*
use ile veritabanı bağlantısı kuruldu.
select* ile bütün sütunlar çağrıldı
from products ile products table'ından veri çektim.
%b ile o ile başlayanlar getirilir not like denildiği zamanda başlamayanların koşulu getirilmiş olur.
*/
```

```sql
/*
(10p) Soru 2 : İsimlerinin ilk harfi a-n arası son harfinin n-z arası ve soyisimlerinin ilk harfi n-z arası son harfinin a-n arazı olan müşterileri listeleyiniz.
               ** Örneğin, İSİM: B......Y ve SOYİSİM: W......D
               * Bütün sütunlar listelenebilir.
*/

use BikeStores
select * from customers
where customers.first_name like '[a-n]%' and customers.first_name like '%[n-z]' 
and customers.last_name like '[n-z]%' and customers.last_name LIKE '%[a-n]' 

/*
use ile veritabanı bağlantısı kurulur
iki harf arasında ki bağlantıyı kurabilmek için iki tırnak arasında köşeli parantez kullanarak 2 harf yazdığın zaman o iki harf arası demek oluyor
dolayısıyla soruda istenen ifadeyi rahat bir şekilde yazabildim.
*/
```

```sql
/*
(20p) Soru 3 : Yöneticileri(staffs.manager_id) ve yöneticilerin yetkili olduğu mağazaları(stores) bulunuz.
               ** Yönetici ID, Yönetici İsmi, Yönetici Soyismi, Mağaza İsmi ve Yöneticiye bağlı olan çalışan sayısı listelenmelidir.
*/

use BikeStores
select staffs.staff_id, staffs.first_name, staffs.last_name, stores.store_name, count(staffs.staff_id) as CalisanSayisi
from stores 
join staffs on stores.store_id=staffs.store_id 
group by stores.store_name, staffs.last_name,staffs.first_name ,staffs.staff_id

/*
Burada ilk başta veritabanı bağlantısı kurulduktan sonra bizden istenen yönetici ID,ismi gibi sütunlar select ile çağrıldı.
from ile stores'den çekildi.
stores ile staffs bağlantısı kurulması için join on kullanıldı.
Bunları gruplayarak bastırmak için group by kullanıldı.
*/
```

```sql
/*
(20p) Soru 4 : Müşterilerin(customers), sipariş(orders and order_items) vermiş olduğu ürünlerin(products) fiyatı 1426'dan büyük olanların kategorilerini bulunuz.
               * Kategori adı ve sayıları listelenmelidir.
*/

use BikeStores
select categories.category_name, count(order_items.quantity) from categories
join products on products.category_id = categories.category_id
join order_items on order_items.product_id = products.product_id
join orders on orders.order_id = order_items.order_id
where products.list_price > 1426
group by categories.category_name

/*
Use ile veritabanı bağlantısı kuruldu.
select ile kategori adı ve sayıları listelendi
group by ile kategori adı gruplandırılarak yazıldı.
bizden istenen ürünlerin fiyatı(products), sipariş (order ve order_items) ile customers birbirine bağlanması gerekmektedir
Bunun için join-on bağlantısı kurulmuştur.
1426 liradan büyük olan kategorileri listelemek için ise where yazılmıştır.
Böylece kod tamamlanmıştır.
*/
```

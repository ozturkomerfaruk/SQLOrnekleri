```sql
/*
(20p) Soru 7: Stok miktarı 19'dan az ve liste fiyatı 2500'den fazla olan ürünler için stok ID, ürün ID, ürün adı, adeti-miktarı ve liste fiyatını listeleyen kodu yazınız.
*/

Use BikeStores

SELECT stocks.store_id, stocks.product_id, products.product_name, stocks.quantity, products.list_price
FROM stocks
JOIN products ON stocks.product_id = products.product_id
WHERE stocks.quantity < 19 AND products.list_price > 2500

/* 
    USE ile database seçildi.
    SELECT ile istenen sütunlar belirlendi.
    FROM hangi table olacağı belirlendi.
    JOIN ile iki table arasında bağ kuruldu.
    WHERE ile istenen koşullar sağlanmış oldu. 
*/
```

```sql
/*
(15p) Soru 8: Adı "Baldwin Bikes" olan mağazada çalışmayan personeller için mağaza adı, şehir, personel adı, personel soyadı, mağaza ID bilgilerini listeleyen kodu yazınız.
*/
    USE BikeStores

    SELECT stores.store_name, stores.city, staffs.first_name, staffs.last_name, staffs.store_id FROM staffs
    JOIN stores ON staffs.store_id = stores.store_id
    WHERE stores.store_name <> 'Baldwin Bikes'

/* 
    Database belirlendi.
    Select ile sütunlar ayarlandı. FROM ile staff table çekildi.
    JOIN ile aradaki bağ kuruldu.
    WHERE ile Baldwin Bikes olmayan mağazalar getirildi.
*/
```

```sql
/*
(15p) Soru 6: Adı Jone olan müşteriye ait tüm siparişler için adı, soyadı, müşteri ID ve sipariş tarihini listeleyen kodu yazınız.
*/
    USE BikeStores

    SELECT customers.customer_id, customers.first_name, customers.last_name,orders.order_date FROM customers
    JOIN orders ON customers.customer_id = orders.customer_id
    WHERE customers.first_name = 'Jone'

/* 
    Database belirlendi. Select ile sütunlar belirlendi ve JOIN ile iki table arasında ki bağ ayarlandı. Where ile Jone isimli şahsın siparişleri geldi.
*/
```

```sql
/*
(15p) Soru 5: Kategori adı "Electric Bikes" olan ürünlerin ürün adı, marka ID, kategori ID ve model bilgilerini listeleyen kodu yazınız.
*/

    USE BikeStores
    
    SELECT product_name, brand_id, categories.category_id, model_year FROM products
    JOIN categories ON products.category_id = categories.category_id WHERE categories.category_name =   'Electric Bikes'

/* 
    SELECT ile sütunlar belirlendi. FROM ile hangi table'dan çekileceği ayarlandı. JOIN ile farklı table ile ile ilişik kurulup WHERE şartı ile sadece Electric Bikes olanlar getirildi.
*/
```

```sql
/*
(10p) Soru 4: Stok adeti-miktarı 7 den büyük 16 dan küçük olan stokları listeleyen kodu yazınız.
*/
    USE BikeStores

    Select * From stocks WHERE quantity > 7 AND quantity < 16

/* 
    Aynı açıklamalarla birlikte WHERE şartı ile birlikte quantity AND operatoru ile koşul durumu sağlanmıştır. 7 ve 16 arası yapılmıştır.
*/
```
